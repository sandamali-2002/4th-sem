```python
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
url = "https://raw.githubusercontent.com/dineshpiyasamara/deep_learning_tutorial/master/04.dropout_regularization/sonar.csv"
df=pd.read_csv(url)
df=df.values
```

```python
df.shape
```

```python
df[:1]
```

```python
X=df[:,0:60].astype(float)
Y=df[:,60]
```

```python
scaler=MinMaxScaler()
X=scaler.fit_transform(X)
```

```python
encoder=LabelEncoder()
encoder.fit(Y)
Y=encoder.transform(Y)
```

```python
from numpy.random.mtrand import rand
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
tf.random.set_seed(42)

model=tf.keras.Sequential([
    tf.keras.layers.Dense(30,activation="relu",input_shape=(60,)),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(60,activation="relu"),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(60,activation="relu"),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(30,activation="relu"),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(30,activation="relu"),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(1,activation="sigmoid")


])
```

```python
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
epoch_number=100
history=model.fit(X_train,y_train,epochs=epoch_number)
```

```python
model.evaluate(X_test,y_test)
```
